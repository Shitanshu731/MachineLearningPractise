# Handling Missing Values for Categorical Datas(String Type)

## Using mode function

### Lets Fill on Column using the mode function

```python
# we get the mode value of a column Named City
# Basically here we get the value with the index
# we just need the value
dataset["City"].mode() 
```

```python
# so , to get the value only we will use
dataset["City"].mode()[0]
```

## Now fill the value in the column "City"

```python
dataset["City"].fillna(dataset["City"].mode()[0],inplace=True)
```

```python
# Get the info about the datas
dataset.info()
```

## Similarly we need to put all the mode values in the respected columns, we should not do it manually , instead we would use For Loop 

```python
# We will get all the columns in object way
dataset.select_dtypes(include="object").columns
```

```python
# Check and Print all the Columns
for i in dataset.select_dtypes(include="object").columns :
    print(i)
```

```python
# in a general way , we are filling all null values with each column's mode value for each column using loop
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
# All the categorical datas missing contents are filled
dataset.isnull().sum()
```
